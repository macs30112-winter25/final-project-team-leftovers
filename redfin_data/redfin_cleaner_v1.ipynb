```python
import sqlite3
import pandas as pd
import json
```

```python
# connect with the db and create cursor object
conn = sqlite3.connect("redfin_properties.db")
cur = conn.cursor()
```

```python
q = """
    SELECT url, data
    FROM properties
    """

json_df = pd.read_sql_query(q, conn)

json_df

conn.close()
```

```python
json_df.info()
```

```python
json_df['data'] = json_df['data'].apply(json.loads)

cleaned_df = pd.json_normalize(json_df['data'])

cleaned_df
```

```python
cleaned_df.info()
```

```python


cleaned_df_dropped_col = cleaned_df.drop(['redfin_agent', 'redfin_agent_average_rating', 'redfin_agent_url', 
                                          'redfin_agent_phone_number', 'redfin_agent_deal_count', 
                                          'redfin_agent_volume', 'redfin_agent_highest_deal',
                                          'agents', 'bike_score', 'walk_score', 'transit', 
                                          'nearby_places', 'image_urls', 'similar_homes', 'url_map_2_x', 
                                          'tax_events', 'notice_content', 'amenities.mls', 'url_map', 'currency',
                                          'description', 'active'
                                          ], axis = 1)

cleaned_df_dropped_col.info()
```

```python
# Remove columns with high non-null
cleaned_df_dropped_col2 = cleaned_df_dropped_col.drop(['amenities.buyers_agent_commission', 'amenities.style', 'amenities.stories', 
                                           'amenities.baths', 'amenities.inventory_type', 'amenities.listing', 'amenities.contact_owner',
                                           'amenities.public_details', 'schools', 'sale_events'
                                          ], axis = 1)

cleaned_df_dropped_col2.info()
```
